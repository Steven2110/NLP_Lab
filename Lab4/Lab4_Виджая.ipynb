```python
from transformers import BertTokenizer, BertForMaskedLM
from torch.nn import functional as F
import torch
```

```python
name = 'bert-base-multilingual-uncased'
tokenizer = BertTokenizer.from_pretrained(name)
model = BertForMaskedLM.from_pretrained(name, return_dict = True)
```

```python
text = "My favourite fruit is " + tokenizer.mask_token + ", mandarin, pear and the fruits name that is used as company name by Steve Jobs."
input =tokenizer.encode_plus(text, return_tensors = "pt")
mask_index = torch.where(input["input_ids"][0] == tokenizer.mask_token_id)
output = model(**input)
```

```python
logits = output.logits
softmax = F.softmax(logits, dim = -1)
mask_word = softmax[0, mask_index[0], :]
top = torch.topk(mask_word, 10)
for i, token in enumerate(top[-1][0].data):
    print(f"{i + 1}. {tokenizer.decode([token])}")
```
